```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#Supervised Learning Tutorial:

X = iris.data
y = iris.target

feature_names = iris.feature_names
target_names = iris.target_names
```

```python
#split data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,y, test_size=0.2) 
#test_size to control the split
print(X_train.shape) #how many rows and columns
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) #how many data points around to observer
knn.fit(X_train,Y_train)
y_pred = knn.predict(X_test)
predicted_species = [iris.target_names[p] for p in y_pred]
```

```python
from sklearn import metrics
print(metrics.accuracy_score(Y_test,y_pred))
```

```python
sample = [[3,5,4,2],[2,3,5,4]]
preds = knn.predict(sample)
pred_species = [iris.target_names[p] for p in preds]
pred_species
```

```python
import joblib

joblib.dump(knn, "iris_model.joblib")
```

```python
model = joblib.load("iris_model.joblib")
iris_s_preds = model.predict(X_test)
preds_species = [iris.target_names[p] for p in iris_s_preds]
print(metrics.accuracy_score(Y_test,iris_s_preds))
```

```python
import seaborn as sns
import pandas as pd
iris_df = pd.DataFrame(X, columns=iris.feature_names)
iris_df = pd.DataFrame(X, columns=iris.feature_names)
iris_df['target'] = y
X_test_df = pd.DataFrame(X_test, columns=iris.feature_names)
X_test_df['predicted'] = iris_s_preds
X_test_df['actual'] = Y_test

sns.scatterplot(
    x=X_test_df['sepal length (cm)'],  
    y=X_test_df['predicted'],
    hue=X_test_df['actual'],  
    palette="viridis"
)
```
